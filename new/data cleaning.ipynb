```python
# Select Important Columns

movies = df[['id','title','overview','genres','keywords']]
movies.head()
```

```python
# Convert Genres to List

import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
```

```python
# Convert Keywords

movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# Convert Overview Into Word List

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# Remove Spaces in Multi-word Genres

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# Create Tags Column

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords']
```

```python
# Convert Tags to String

movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
```

```python
# Final Dataset for Recommendation

new_df = movies[['id','title','tags']]
new_df.head()
```

```python
# Apply Stemming 
# Stemming reduces words to their root form. (here. using NLTK)

# Example:
# loved → love  
# loving → love

import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from collections import Counter

genre_list = []

for g in df['genres']:
    for genre in g.split(','):
        genre_list.append(genre.strip())

genre_counts = Counter(genre_list)

genre_df = pd.DataFrame(genre_counts.items(), columns=['Genre', 'Movie_Count'])
genre_df = genre_df.sort_values(by='Movie_Count', ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x='Movie_Count', y='Genre', data=genre_df.head(15))

plt.title("Number of Movies per Genre")
plt.xlabel("Number of Movies")
plt.ylabel("Genre")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Create a copy to keep your original dataframe intact
df_genre = df.copy()

# 2. Convert 'vote_average' to a 'rating' column for easier use
df_genre['rating'] = df_genre['vote_average']

# 3. Clean and Split genres
# We use .astype(str) to ensure split() works even if there are missing values
df_genre['genres'] = df_genre['genres'].fillna('').astype(str).apply(lambda x: x.split(','))

# 4. Convert each genre into a separate row
df_genre = df_genre.explode('genres')

# 5. Remove extra spaces and handle empty strings from the split
df_genre['genres'] = df_genre['genres'].str.strip()
df_genre = df_genre[df_genre['genres'] != '']

# 6. Calculate average rating per genre
# Note: Use lowercase 'genres' and 'rating' to match the columns we just made
genre_rating = df_genre.groupby('genres')['rating'].mean().reset_index()

# 7. Sort values
genre_rating = genre_rating.sort_values(by='rating', ascending=False)

# 8. Plot
plt.figure(figsize=(12, 8))
# Plotting the top 15 genres
sns.barplot(x='rating', y='genres', data=genre_rating.head(15), palette='viridis')

plt.title("Top 15 Average Ratings per Genre", fontsize=15)
plt.xlabel("Average Rating (Out of 10)", fontsize=12)
plt.ylabel("Genre", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Create a copy and handle missing values
df_counts = df.copy()
df_counts['genres'] = df_counts['genres'].fillna('Unknown')

# 2. Split the strings into lists and explode them into individual rows
# This turns "Action, Comedy" into two separate entries for the count
df_counts['genres'] = df_counts['genres'].astype(str).str.split(',')
df_exploded = df_counts.explode('genres')

# 3. Clean up whitespace
df_exploded['genres'] = df_exploded['genres'].str.strip()

# 4. Count occurrences of each genre
genre_counts = df_exploded['genres'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Movie_Count']

# 5. Plotting
plt.figure(figsize=(12, 8))
sns.barplot(
    x='Movie_Count', 
    y='Genre', 
    data=genre_counts.head(20), # Showing top 20 genres
    palette='magma'
)

plt.title("Number of Movies per Genre", fontsize=15)
plt.xlabel("Count of Movies", fontsize=12)
plt.ylabel("Genre", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.show()
```
